```python
# load dataset
import pandas as pd

df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_394176.csv', skiprows = 4)
df.head()
```

```python
# selecting only a single year, 2021
df_2021 = df[['Country Name', '2021']].dropna()
df_2021.columns = ['Country', 'Population']
```

```python
# select top 15 countries only
top15 = df_2021.sort_values(by='Population', ascending=False).head(15)
```

```python
# visualization
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (8, 6))
sns.barplot(x='Country', y='Population', data=top15)
plt.title('Bar Chart of Top 15 Countries in 2021')
plt.xlabel('Countries')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# histogram for all countries
plt.figure(figsize=(8, 5))
sns.histplot(df_2021['Population'], bins=30,kde=True, color='green')

plt.title('Distribution of Population Across Countries (2021)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```
